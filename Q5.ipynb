بارگذاری داده تمیز

```python
import pandas as pd
from pathlib import Path

# مسیر پروژه = یک پوشه بالاتر از محل اجرای نوت‌بوک
project_root = Path.cwd().parent

# مسیر نسبی به processed/v2 و reports/answers
p = project_root / "data" / "processed" / "v2"
ans_dir = project_root / "reports" / "answers"
ans_dir.mkdir(parents=True, exist_ok=True)

# بارگذاری داده
df_match = pd.read_parquet(p / "match_clean_v2.parquet")
print("Shape:", df_match.shape)
```

ستون ها 

```python
set_cols = [c for c in df_match.columns if c.startswith("period_")]
print("Detected set columns:", set_cols[:10])
```

شمارش ست ها 

```python
# تعداد ست‌های معتبر (غیر صفر و غیر NaN) برای هر مسابقه
df_match["num_sets"] = df_match[set_cols].notna().sum(axis=1)

# بررسی نمونه
print(df_match[["match_id", "num_sets"]].head(10))
```

میانگین ست ها 

```python
avg_sets = df_match["num_sets"].mean()

print(" Average sets per match:", round(avg_sets, 2))
```

ذخیره نتیجه

```python
out_path = ans_dir / "q05_avg_sets.csv"
pd.DataFrame([{"average_sets_per_match": round(avg_sets, 2)}]).to_csv(out_path, index=False)
print("Saved to", out_path)
```
