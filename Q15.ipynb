Question 15. How many distinct countries are represented in the dataset? 

```python
import pandas as pd
from pathlib import Path

project_root = Path.cwd().parent
raw_dir = project_root / "data" / "raw"
answers_dir = project_root / "reports" / "answers"
answers_dir.mkdir(parents=True, exist_ok=True)

players_file = raw_dir / "players.parquet"
players = pd.read_parquet(players_file)

distinct_count = players["country"].dropna().nunique()

output_file = answers_dir / "q15_distinct_countries.csv"
pd.DataFrame({"distinct_countries": [distinct_count]}).to_csv(
    output_file, index=False, encoding="utf-8-sig"
)

print(f"Total of {distinct_count} distinct countries are represented in the dataset.")
print(f"Saved to {output_file}")
```
