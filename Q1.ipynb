 1. How many tennis players are included in the dataset? 


```python
import pandas as pd
from pathlib import Path

project_root = Path.cwd().parent
raw_dir = project_root / "data" / "raw"
out_dir = project_root / "reports" / "answers"
out_dir.mkdir(parents=True, exist_ok=True)

players_path = raw_dir / "players.parquet"
df_players = pd.read_parquet(players_path, engine="pyarrow")

unique_count = df_players["player_id"].nunique()
print(f"Total unique players: {unique_count}")

output_path = out_dir / "q1_unique_players.csv"
pd.DataFrame({"total_unique_players": [unique_count]}).to_csv(
    output_path, index=False, encoding="utf-8-sig"
)
print(f"Saved to {output_path}")
```
